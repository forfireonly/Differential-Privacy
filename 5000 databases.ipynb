how does torch.rand(num_entries) > 0.5 work?
torch.rand(num_entries) returns values between 0 and 1
> returns false(0) or true(1),  so see the example.

```python
b=torch.rand(4)
```

```python
print(b)
```

```python
i=0
databases= list()
num_entries = 5000
while i < 5000:
    db = torch.rand(num_entries) > 0.5
    db = torch.cat((db[0:i], db[i+1:]))
    databases.append(db)
    i+=1    
```

```python
print(databases)
```

```python
len(databases)
```

```python
len(databases[3])
```

```python
db1 = torch.rand(num_entries) > 0.5
```

```python
def removeEntry(db, entry):
    return torch.cat((db1[0:entry], db1[entry+1:]))
```

```python
parallel_databases = list()
for i in range(len(databases)):
    pdb = removeEntry(db, i)
    parallel_databases.append(pdb)
    
```
